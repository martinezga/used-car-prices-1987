### Identifying and handling missing values

```python
df.head(30)
```

In the car dataset, missing data comes with the question mark "?". We replace "?" with NaN (Not a Number), which is Python's default missing value marker, for reasons of computational speed and convenience.

```python
import numpy as np

# replace "?" to NaN
df.replace("?", np.nan, inplace = True)
df.head(15)
```

```python
missingData = df.isnull()
missingData.head()
```

The output is a boolean value. "True" stands for missing value, while "False" stands for not missing value.

```python
for column in missingData.columns.values.tolist():
    print(column)
    print(missingData[column].value_counts())
    print('')
```

**Replace by mean:**

* "normalized-losses": 41 missing data
* "stroke": 4 missing data
* "bore": 4 missing data
* "horsepower": 2 missing data
* "peak-rpm": 2 missing data

**Replace by frequency:**

* "num-of-doors": 2 missing data, replace them with "four".
    - Reason: 84% sedans is four doors. Since four doors is most frequent, it is most likely to occur

**Drop the whole row:**

* "price": 4 missing data
    - Reason: price is what we want to predict. Any data entry without price data cannot be used for prediction; therefore any row now without price data is not useful.

```python
avgNormLoss = df['normalized-losses'].astype('float').mean(axis=0)
print('Average of normalized-losses:', avgNormLoss)
```

```python
df['normalized-losses'].replace(np.nan, avgNormLoss, inplace=True)
```

```python
avgStroke = df['stroke'].astype('float').mean(axis=0)
print('Average of stroke:', avgStroke)
```

```python
df['stroke'].replace(np.nan, avgStroke, inplace=True)
```

```python
avgBore = df['bore'].astype('float').mean(axis=0)
print('Average of bore:', avgBore)
```

```python
df['bore'].replace(np.nan, avgBore, inplace=True)
```

```python
avgHorseP = df['horsepower'].astype('float').mean(axis=0)
print(avgHorseP)
```

```python
df['horsepower'].replace(np.nan, avgHorseP, inplace=True)
```

```python
avgPeakRpm = df['peak-rpm'].astype('float').mean(axis=0)
print(avgPeakRpm)
```

```python
df['peak-rpm'].replace(np.nan, avgPeakRpm, inplace=True)
```

```python
df['num-of-doors'].value_counts()
```

```python
df['num-of-doors'].replace(np.nan, 'four', inplace=True)
```

```python
# simply drop whole row with NaN in "price" column
df.dropna(subset=['price'], axis=0, inplace=True)

# reset index, because we droped rows
df.reset_index(drop=True, inplace=True)
```

### Data transformation

Fuel consumption columns "city-mpg" and "highway-mpg" are represented by mpg (miles per gallon) unit. Assume we are developing an application in a country that accept the fuel consumption with L/100km standard.

We will need to apply data transformation to transform mpg into L/100km.

The formula for unit conversion is

L/100km = 235 / mpg

```python
df['city-L/100km'] = 235/df["city-mpg"]
```

```python
df.head()
```

```python
df['highway-mpg'] = 235/df['highway-mpg']
df.rename(columns={'highway-mpg':'highway-L/100km'}, inplace=True)
df.head()
```

```python
df[['city-mpg', 'highway-L/100km', 'city-L/100km']] = df[['city-mpg', 'highway-L/100km', 'city-L/100km']].astype("float")
```

```python
df.dtypes
```

### Binning

We will use the Pandas method 'cut' to segment the 'horsepower' column into 3 bins

```python
df["horsepower"]=df["horsepower"].astype(int, copy=True)
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df["horsepower"])

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
```

```python
group_names = ['Low', 'Medium', 'High']
```

```python
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True )
df[['horsepower','horsepower-binned']].head(20)
```

```python
df["horsepower-binned"].value_counts()
```

```python
pyplot.bar(group_names, df["horsepower-binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
# draw historgram of attribute "horsepower" with bins = 3
plt.pyplot.hist(df["horsepower"], bins = 3)

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

### Indicator variable (or dummy variable)

```python
dummyVar1 = pd.get_dummies(df["fuel-type"])
dummyVar1.head()
```

```python
dummyVar1.rename(columns={'diesel':'fuel-type-diesel', 'gas':'fuel-type-gas'}, inplace=True)
dummyVar1.head()
```

```python
# merge data frame
df = pd.concat([df, dummyVar1], axis=1)

# drop original column "fuel-type"
df.drop('fuel-type', axis = 1, inplace=True)
```

```python
df.columns
```

```python
dummyVar2 = pd.get_dummies(df['aspiration'])
dummyVar2.head()
```

```python
dummyVar2.rename(columns={'std': 'aspiration-std', 'turbo': 'aspiration-turbo'}, inplace=True)
dummyVar2.head()
```

```python
df = pd.concat([df, dummyVar2], axis=1)
df.drop('aspiration', axis=1, inplace=True)
```

Scale the columns "length", "width" and "height"
